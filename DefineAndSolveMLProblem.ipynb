```python
correlation_matrix = df.corr()
top_features = correlation_matrix['Happiness'].abs().sort_values(ascending=False).index[1:3]
print(top_features)
```

```python
for feature in top_features:
    sns.scatterplot(x=df[feature], y=df['Happiness'])
    plt.xlabel(feature)
    plt.ylabel('Happiness')
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.regplot(x='LogGDP', y='Happiness', data=df);
```

```python
y = df['Happiness']
features = {'LogGDP',
       'SocialSupport', 'LifeExpectancy',
       'Freedom', 'Generosity',
       'Corruption', 'Positive', 'Negative',
       'GovtConfidence', 'DemocraticQuality',
       'DeliveryQuality', 'Gini', 'Gini15'}
X = df[list(features)]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_lr_pred = lr_model.predict(X_test)
```

```python
lr_rmse = mean_squared_error(y_test, y_lr_pred, squared = False)
lr_r2 = r2_score(y_test, y_lr_pred)
print('[LR] Root Mean Squared Error: {0}'.format(lr_rmse))
print('[LR] R2: {0}'.format(lr_r2))
```

```python
print('Intercept:')
print('alpha = ' , lr_model.intercept_)
print('Weights:')
i = 0
for w in lr_model.coef_:
    print('w_', i+1,'= ', w, ' [ weight of', list(features)[i],']')
    i += 1
```

Gini, Positive, and SocialSupport have the largest weights. This means that increasing those features would directly correspond to an increase in Happiness. According to these weights, people tend to be happier when they have a higher life expectancy, have more support, have better quality of delivery, have more generosity, have more freedom, have a higher rate of GDP, and have more positive emotions. Gini15, GovtConfidence, Corruption, and Democratic Quality all have negative weights. This means that decreasing those features would directly correspond to an increase in Happiness. Lower levels of corruption and less negative emotions contribute to increased happiness as well. This reflects the idea that more positive societal conditions generally correlate with greater overall well-being.

An interesting feature in the data is the World Bank estimate of the Gini index. The Gini and Gini15 have the highest and lowest weights, respectively, showing that the Gini index in general plays a crucial role in capturing the impact of income inequality on happiness. A lower Gini coefficient, indicating more equality, can positively impact happiness. A higher Gini coefficient, indicating more inequality, can negatively impact happiness.

Now, let's move on to testing other models.

```python
dt_regressor = DecisionTreeRegressor(random_state=1234)
```

```python
param_grid = {
    'max_depth': [2, 4, 6, 8, 10, 12, 14],
    'min_samples_leaf': [5, 10, 15, 20, 25, 30, 35, 40, 45, 50],
    'min_samples_split': [2, 5, 10, 15, 20],
    'max_features': ['sqrt', 'log2', None]
}
```

```python
dt_regressor = DecisionTreeRegressor(random_state = 1234)
dt_grid = GridSearchCV(estimator = dt_regressor, param_grid = param_grid, cv = 3, scoring = 'neg_root_mean_squared_error')
dt_grid_search = dt_grid.fit(X_train, y_train)
```

```python
rmse_DT = -1 * dt_grid_search.best_score_
print("[DT] RMSE for the best model is : {:.2f}".format(rmse_DT) )
```

```python
dt_best_params = dt_grid_search.best_params_

dt_best_params
```

```python
best_max_depth = dt_best_params['max_depth']
best_min_samples_leaf = dt_best_params['min_samples_leaf']

dt_model = DecisionTreeRegressor(max_depth = best_max_depth, min_samples_leaf = best_min_samples_leaf, random_state = 1234)

dt_model.fit(X_train, y_train)
```

```python
y_dt_pred = dt_model.predict(X_test)
dt_mse = mean_squared_error(y_test, y_dt_pred)
dt_rmse = np.sqrt(dt_mse)
dt_r2 = r2_score(y_test, y_dt_pred)
print('[DT] Root Mean Squared Error: {0}'.format(dt_rmse))
print('[DT] R2: {0}'.format(dt_r2))
```

```python
estimators = [("DT", DecisionTreeRegressor(max_depth=8, min_samples_leaf=25)),
              ("LR", LinearRegression())
             ]
```

```python
stacking_model = StackingRegressor(
    estimators = estimators,
    passthrough = False
)
stacking_model.fit(X_train, y_train)
```

```python
stacking_pred = stacking_model.predict(X_test)

stack_mse = mean_squared_error(y_test, stacking_pred)
stack_rmse = np.sqrt(stack_mse)

stack_r2 = r2_score(y_test, stacking_pred)
   
print('Root Mean Squared Error: {0}'.format(stack_rmse))
print('R2: {0}'.format(stack_r2))
```

```python
gbdt_model = GradientBoostingRegressor(max_depth = 2, n_estimators = 300, random_state = 1234)
gbdt_model.fit(X_train, y_train)
```

```python
y_gbdt_pred = gbdt_model.predict(X_test)

gbdt_mse = mean_squared_error(y_test, y_gbdt_pred)
gbdt_rmse = np.sqrt(gbdt_mse)

gbdt_r2 = r2_score(y_test, y_gbdt_pred)

print('[GBDT] Root Mean Squared Error: {0}'.format(gbdt_rmse))
print('[GBDT] R2: {0}'.format(gbdt_r2))
```

```python
importances = gbdt_model.feature_importances_
feature_importances = pd.DataFrame({'Feature': list(features), 'Importance': importances}).sort_values(by='Importance', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importances)
plt.title('Feature Importances from Gradient Boosting Decision Trees')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
rf_model = RandomForestRegressor(max_depth = 32, n_estimators = 300, random_state = 1234)
rf_model.fit(X_train, y_train)
```

```python
y_rf_pred = rf_model.predict(X_test)
rf_mse = mean_squared_error(y_test, y_rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, y_rf_pred)
print('[RF] Root Mean Squared Error: {0}'.format(rf_rmse))
print('[RF] R2: {0}'.format(rf_r2))    
```

```python
importances = rf_model.feature_importances_
feature_importances = pd.DataFrame({'Feature': list(features), 'Importance': importances}).sort_values(by='Importance', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importances)
plt.title('Feature Importances from Random Forest')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
RMSE_Results = [stack_rmse, lr_rmse, dt_rmse, gbdt_rmse, rf_rmse]
R2_Results = [stack_r2, lr_r2, dt_r2, gbdt_r2, rf_r2]
rg= np.arange(5)
width = 0.35
plt.bar(rg - width/2, RMSE_Results, width, label = 'RMSE', color = 'b')
plt.bar(rg + width/2, R2_Results, width, label = 'R2', color = 'g')
labels = ['Stacking','LR', 'DT', 'GBDT', 'RF']
plt.xticks(rg, labels)
plt.xlabel("Models")
plt.ylabel("RMSE/R2")
plt.ylim([0,1])
plt.title('Model Performance')
plt.legend(loc='upper left', ncol=2)
plt.show()
```

This bar chart compares the performance of Stacking, Linear Regression (LR), Decision Tree (DT), Gradient Boosting Decision Tree (GBDT), and Random Forest (RF)—using two key metrics: Root Mean Square Error (RMSE) and R-squared ($R^2$). The blue bars represent RMSE, where lower values indicate better performance due to lower prediction errors. The green bars represent $R^2$, where higher values signify better performance by explaining a greater proportion of variance in the data. Among the models, the Random Forest (RF) stands out with the lowest RMSE and the highest $R^2$ scores, demonstrating its superior accuracy and predictive power compared to the others. This indicates that the Random Forest model provides the most accurate predictions with the least error, making it the most reliable choice for the World Happiness Report dataset.
